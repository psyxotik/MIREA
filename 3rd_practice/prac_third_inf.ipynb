## Задача:
    Найти представление беззнакового числа 132 в 10 в шестнадцатиразрядном
    битовом наборе

```python
n = 132
s = ''
while n > 0:
    s = str(n % 2) + s
    n //= 2
print(s.zfill(16))
```
